## 1. 막대 그래프

```python
from matplotlib import pyplot as plt
%matplotlib inline

# 그래프 데이터 
subject = ['English', 'Math', 'Korean', 'Science', 'Computer']
points = [40, 90, 50, 60, 100]

# 그래프 그리기
plt.bar(subject, points)

plt.xlabel("subject")
plt.ylabel("points")
plt.title("Yuna's Test Result")
plt.show()
```

## 2. 선 그래프

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

# 그래프 데이터 
x_values = np.arange(5)
y_values = np.arange(5)*2

# 그래프 그리기
fig = plt.figure(figsize=(2,4))

plt.plot(x_values, y_values)
plt.xlabel("x")
plt.ylabel("y")
plt.xticks(np.arange(5))
plt.title("y = 2x")
plt.show()
```

## 3. 산점도 그래프

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

# 그래프 데이터
x_values = np.arange(5)
y_values = np.arange(5)*2

# 그래프 그리기
fig = plt.figure(figsize=(2,4))

plt.scatter(x_values, y_values)
plt.xlabel("x")
plt.ylabel("y")
plt.xticks(np.arange(5))
plt.title("y = 2x")
plt.show()
```

## 4. 원 그래프

```python
from matplotlib import pyplot as plt
%matplotlib inline

# 그래프 데이터
data = [340, 320, 160, 180]
labels = ["Apple", "Banana", "Melon", "Grapes"]
explode = [0.05, 0.05, 0.05, 0.05]

# 그래프 그리기
plt.pie(data, labels=labels, autopct='%.1f%%', explode=explode)
plt.title("pie graph")
plt.show()
```

## 5. subplot

```python
from matplotlib import pyplot as plt
%matplotlib inline


fig = plt.figure(figsize=(10, 14))

# 첫번째 그래프
subject = ['English', 'Math', 'Korean', 'Science', 'Computer']
points = [40, 90, 50, 60, 100]

plt.subplot(2, 2, 1)
plt.bar(subject, points)
plt.xlabel("subject")
plt.ylabel("points")
plt.title("Yuna's Test Result")

# 두번째 그래프
x_values = np.arange(5)
y_values = np.arange(5)*2

plt.subplot(2, 2, 2)
plt.plot(x_values, y_values)
plt.xlabel("x")
plt.ylabel("y")
plt.xticks(np.arange(9))
plt.title("y = 2x")

# 세번째 그래프
x_values = np.arange(5)
y_values = np.arange(5)*2

plt.subplot(2, 2, 3)
plt.scatter(x_values, y_values)
plt.xlabel("x")
plt.ylabel("y")
plt.xticks(np.arange(9))
plt.title("y = 2x")

# 네번째 그래프
data = [340, 320, 160, 180]
labels = ["Apple", "Banana", "Melon", "Grapes"]
explode = [0.05, 0.05, 0.05, 0.05]

plt.subplot(2, 2, 4)
plt.pie(data, labels=labels, autopct='%.1f%%', explode=explode)
plt.title("pie graph")

plt.suptitle('Subplots') # 전체 타이틀 추가
plt.show()
```
